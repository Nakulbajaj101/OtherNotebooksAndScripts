```python
sns.set_style("whitegrid")
g = sns.JointGrid(x="MSRP",
            y="SALES",
            data=data,
                 xlim = (50,230)) 

g.plot(sns.regplot, sns.distplot)

plt.show()
plt.clf()
```

```python
sns.jointplot(x="MSRP",
            y="SALES",
            data=data, xlim = (50,230), kind='reg') 
plt.show()
plt.clf()
```

```python
sns.jointplot(x="MSRP",
            y="SALES",
            data=data, xlim = (50,230), kind='reg', order = 3) 
plt.show()
plt.clf()
```

```python
sns.jointplot(x="MSRP",
            y="SALES",
            data=data, xlim = (50,230), kind='resid', order = 3) 
plt.show()
plt.clf()
```

```python
g = (sns.jointplot(x="MSRP",
             y="SALES",
             kind='scatter',
             data=data,
             marginal_kws=dict(bins=15, rug=True))
    .plot_joint(sns.kdeplot))
g.annotate(stats.pearsonr)
    
plt.show()
plt.clf()
```
